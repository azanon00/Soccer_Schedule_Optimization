```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Data201819SerieA.csv')
data.head()
```

```python
# remove unnecessary columns
data = data[["Date", "HomeTeam", "AwayTeam"]]
```

```python
# map each team to a number
teams = data["HomeTeam"].unique()
```

```python
data.shape
```

There are 380 rows because of 380 matches in total. We optimize the first half, thus letâ€™s take only the first 190 rows.

```python
data = data.iloc[:190, :]
data.shape
```

```python
# for each date, compute which day it is in our time horizon
import datetime

format = "%d/%m/%Y"

data["Date"] = data["Date"].apply(lambda x: datetime.datetime.strptime(x, format))

# find first day in horizong
first_day = data["Date"].min()
```

```python
# compute number of days since first day
data["Day"] = data["Date"].apply(lambda x: (x - first_day).days)
```

From data, get day in which each team plays (independently of whether it's home or away)

```python
# create matrix of 20 rows (1 per team) and 134 columns (1 per day). Cell is 1 if team plays on that day, 0 otherwise
matrix = np.zeros((20, 134))
```

```python
matrix.shape
```

```python
data.loc[189, "Day"]
```

```python
for team in teams:
    for i, (home, away) in enumerate(zip(data["HomeTeam"], data["AwayTeam"])):
        if home == team or away == team:
            matrix[teams.tolist().index(team), data.loc[i, "Day"]] = 1
```

```python
pd.DataFrame(matrix)
```

```python
def GetC():

    W_ = 19
    D = 7

    C = np.zeros((D*(W_+1), D*(W_+1)))

    for t in range(W_*D+1):
        for t_ in range(W_*D+1):
            if t < t_:
                C[t, t_] = (t_ - t)**(-1)

    return C
```

```python
GetC()
```

```python
# define total cost for each team
team_total_cost = []

# for each team 
for i in range(1, 20):
    # find index of nonzer elements, they correspond to the days in which the team plays
    team_plays = np.where(matrix[i, :] != 0)[0]

    # we will find W non-zero elements, since each team plays against all the other teams
    # for i = 1:W, return GetC()[team_plays[i], team_plays[i+1]] and sum
    team_total_cost = np.append(team_total_cost, sum([GetC()[team_plays[i], team_plays[i+1]] for i in range(1,19-1)]))


#Â total cost of the calendar
objective = sum(team_total_cost)
```
